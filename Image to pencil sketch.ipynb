# Import required modules

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r"C:\Users\sameer narwade\Desktop\flowers.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,7))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# Converting Image to Grayscale

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(image_gray,cmap="gray")
plt.title("GrayScale Image")
plt.axis("off")
```

```python
image_invert = cv2.bitwise_not(image_gray)
plt.figure(figsize=(7,7))
plt.imshow(image_invert,cmap="gray")
plt.title("Inverted Image")
plt.axis("off")
```

```python
img_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(7,7))
plt.imshow(img_smoothing,cmap="gray")
plt.title("Smoothen Image")
plt.axis("off")
```

# Converting Image to Pencil sketche

```python
final = cv2.divide(image_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,7))
plt.imshow(final,cmap="gray")
plt.title("Pencil Sketch Image")
plt.axis("off")
plt.show()
```
